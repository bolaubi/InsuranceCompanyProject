```python
def matrix_checker(df):
    matrix_checker = []
    for i in df.columns:
        matrix_checker.append([
            df[i].dtype,
            df[i].isna().sum(),
            f'{round(df[i].isna().sum()/len(df[i])*100,2)}%',
            df[i].nunique(),
            df[i].unique()[:8]
        ])
    temp_df = pd.DataFrame(matrix_checker, index=df.columns.values, columns='Dtype Total_Nan Nan_Pct Num_Unique Example'.split())
    return temp_df
```

```python
# Creating Feature Engineered DF 1
def createDFEngineered_1(df):
    global tqdm
    global cust_idxValues
    
    customer_lifetime = []  # Number of years as our customer
    isLoyal = []  # Customer who has spent more than a year purchasing our insurance products
    premium_lifetime = []  # Total premium during the customer's tenure with us
    purchase_frequency = []  # Total purchase frequency of customers
    
    for i, j in tqdm(cust_idxValues.items()):
        if j==1:
            customer_lifetime.append(0)
            isLoyal.append('0')
            premium_lifetime.append(df[df['Customer ID']==i]['Premium Amount'].values[0])
            purchase_frequency.append(j)
            
        else: 
            lifetime = (df[df['Customer ID']==i]['Age'].iloc[0])-(df[df['Customer ID']==i]['Age'].iloc[-1])
            if lifetime == 0:
                customer_lifetime.append(0)
                isLoyal.append('0')
                premium_lifetime.append(df[df['Customer ID']==i]['Premium Amount'].sum())
                purchase_frequency.append(j)
            else: 
                customer_lifetime.append(lifetime)
                isLoyal.append('1')
                premium_lifetime.append(df[df['Customer ID']==i]['Premium Amount'].sum())
                purchase_frequency.append(j)

    # Creating Feature Engineered DF that is going to be added to the main DF
    matrix_FE = [cust_idx, customer_lifetime, isLoyal, premium_lifetime, purchase_frequency]
    df_FE = pd.DataFrame(matrix_FE).T
    df_FE.columns = ['Customer ID', 'customer_lifetime', 'isLoyal', 'premium_lifetime', 'purchase_frequency']
     
    return df_FE
```

```python
df_FE1 = createDFEngineered_1(df)
```

```python
# Creating a function to return products bought by the user and their scores
# Each product starts with a score of 0, then deducts 1 for a response of '0' and adds 1 for a response of '1'

# Note: adding a function to sum up all of the lifetime premium account based on 'Response' variable ('1' and '0')
def userProductScoring(custID_list):
    global tqdm
    userProd_result = []  # Products purchased by customers and their corresponding scores assigned by customers
    premiumYes = []  # Total lifetime premium for customers with a 'Response' value of 1.
    premiumNo = []  # Total lifetime premium for customers with a 'Response' value of 0.

    def add_scoringProd(list_of_tuples):
        container = {}
        for i in list_of_tuples:
            if i[0] not in container:
                container[i[0]] = i[1] * 1.5  # The most recent experience with the product carries higher importance or weight
            else:
                container[i[0]] += i[1]
        return container
    
    def add_scoring(list_of_tuples):
        container = {}
        for i in list_of_tuples:
            if i[0] not in container:
                container[i[0]] = i[1]
            else:
                container[i[0]] += i[1]
        return container

    for i in tqdm(custID_list):
        planCode_ = list(df[df['Customer ID']==i].loc[:, 'Plan Code'])
        scoring_ = list(df[df['Customer ID']==i].loc[:, 'Scoring'])
        list_PC_scoring = list(zip(planCode_, scoring_))  # List 'Plan Code' & 'scoring'
        userProd_result.append([i, add_scoringProd(list_PC_scoring)])
        
        response_cust = list(df[df['Customer ID']==i].loc[:, 'Response'])
        premium_cust = list(df[df['Customer ID']==i].loc[:, 'Premium Amount'])
        list_RP = list(zip(response_cust, premium_cust))
        
        premiumYesNo = add_scoring(list_RP)
        if ('1' in premiumYesNo) and ('0' not in premiumYesNo):
            premiumYes.append(add_scoring(list_RP)['1'])
            premiumNo.append(0)
        elif ('0' in premiumYesNo) and ('1' not in premiumYesNo):
            premiumYes.append(0)
            premiumNo.append(add_scoring(list_RP)['0'])
        else:
            premiumYes.append(add_scoring(list_RP)['1'])
            premiumNo.append(add_scoring(list_RP)['0'])
        
        
    return userProd_result, premiumYes, premiumNo
```

```python
def createDFEngineered_2():
    global cust_idxValues
    idxValuesAbove1 = [i for i,j in cust_idxValues.items() if j>1]
    idxValuesRemainder = [i for i in cust_idx if i not in idxValuesAbove1]
    
    matrix_idxValuesAbove1, list_PYesAbove1, list_PNoAbove1 = userProductScoring(idxValuesAbove1)
    matrix_idxValuesRemainder, list_PYesRemainder, list_PNoRemainder  = userProductScoring(idxValuesRemainder)
    
    matrix_userProdScoreCom = np.concatenate([matrix_idxValuesAbove1, matrix_idxValuesRemainder], axis=0)
    df_FE2 = pd.DataFrame(matrix_userProdScoreCom, columns=['Customer ID', 'prodScore'])
    df_FE2['premium_LTYes'] = list_PYesAbove1+list_PYesRemainder  # joining the list 'premiumYes'
    df_FE2['premium_LTNo'] = list_PNoAbove1+list_PNoRemainder  # joining the list 'premiumNo'
    
    return df_FE2
```

```python
# Creating Shapiro-Wilk Normality Test 
def ShapiNormalTest(colname, alpha):
    statistic, pvalue = ss.shapiro(df[colname])

    if pvalue < alpha:
        print()
        print(f"H1: With {1-alpha} confidence level, '{colname}' DOES 'NOT' COME from normally distributed data")
    else: 
        print()
        print(f"H0: With {1-alpha} confidence level, '{colname}' DOES COME from normally distributed data")
    print(f"The statistic is: {round(statistic, 3)}, and the Pvalue: {pvalue}")
```

```python
for i in list_num:
    ShapiNormalTest(i, 0.05)
```

The Shapiro-wilk's normality test supports previous claim that numerical columns are **not normally distributed**.

```python
# Creating an outlier checker function
def outlierChecker(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3-q1
    lower_whisker = q1-1.5*iqr
    upper_whisker = q3+1.5*iqr
    arr_outlier = np.array([i for i in series if (i < lower_whisker) or (i > upper_whisker)])
    return arr_outlier

# Creating dictionary outlier checker function
def dict_outlier(list_):
    list_tupNumerical = [(i, outlierChecker(df[i])) for i in list_]
    dict_Numerical = {i:j for (i,j) in list_tupNumerical}
    return dict_Numerical
```

> 02.05c: Categorical Features - further preprocessing

```python
# df[(df['Married']=='No') & (df['Insurance Type']=='Joint')]  # output: 412 rows --> encoded as 1
# df[(df['Married']=='No') & (df['Insurance Type']=='Personal')]  # output: 11963 rows --> encoded as 2
# df[(df['Married']=='Yes') & (df['Insurance Type']=='Joint')]  # output: 2813 rows --> encoded as 3
# df[(df['Married']=='Yes') & (df['Insurance Type']=='Personal')]  # output: 0 rows --> encoded as 0
```

```python
# Merging 'Married' and 'Insurance Type' columns into a new feature called 'Married_Insured'.
married_insured = np.where(
    (df['Married']=='No') & (df['Insurance Type']=='Joint'), '1', 
        np.where((df['Married']=='No') & (df['Insurance Type']=='Personal'), '2', 
                np.where((df['Married']=='Yes') & (df['Insurance Type']=='Joint'), '3', '0')))

df.insert((df.columns.get_loc('Married')+1), 'Married_Insured', married_insured)
list_cat.insert((df.select_dtypes('object').columns.get_loc('Married_Insured')), 'Married_Insured')
# df[df['Married_Insured']==3]['Insurance Type'].value_counts()  # Checking
```

```python
# Dropping 'Region Code', 'Married' and 'Insurance Type' columns
cat_toRemove = ['Region Code', 'Married', 'Insurance Type']
df.drop(columns=cat_toRemove, inplace=True)
list_cat = [i for i in list_cat if i not in cat_toRemove]
```
